```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Iris dataset
iris = sns.load_dataset('iris')

# Extract sepal length data
sepal_length = iris['sepal_length']

# Plot histogram using Seaborn
sns.histplot(sepal_length, kde=True) # kde=True for the KDE curve
plt.title('Histogram of Sepal Length in Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
from scipy import stats
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()

# Extract sepal length data
sepal_length = iris.data[:, 0]

# Calculate mean
mean_sepal_length = np.mean(sepal_length)

# Calculate median
median_sepal_length = np.median(sepal_length)

# Calculate mode
mode_sepal_length = stats.mode(sepal_length)

# Calculate standard deviation
std_sepal_length = np.std(sepal_length)

print("Mean Sepal Length:", mean_sepal_length)
print("Median Sepal Length:", median_sepal_length)
print("Mode sepal Length:", mode_sepal_length.mode)
print("Standard Deviation of Sepal Length:", std_sepal_length)
```
