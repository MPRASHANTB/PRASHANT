```python
#Q4

import numpy as np
import matplotlib.pyplot as plt

# Set the random seed
np.random.seed(27)

# Randomly select coefficients for a 5th degree polynomial
coefficients = np.random.randint(6, 10, size=6)

# Define the polynomial function
def polynomial(x):
    return np.polyval(coefficients, x)

# Generate x values
x_values = np.linspace(-5, 5, 100)

# Compute y values using the polynomial function
y_values = polynomial(x_values)

# Plot the polynomial
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Polynomial')

# Find maxima and minima of the polynomial
maxima_x = x_values[np.argmax(y_values)]
minima_x = x_values[np.argmin(y_values)]

# Annotate maxima and minima
plt.scatter([maxima_x, minima_x], [polynomial(maxima_x), polynomial(minima_x)], color='red')
plt.annotate('Maxima', xy=(maxima_x, polynomial(maxima_x)), xytext=(maxima_x - 2, polynomial(maxima_x) + 50),
             arrowprops=dict(facecolor='red', shrink=0.05))
plt.annotate('Minima', xy=(minima_x, polynomial(minima_x)), xytext=(minima_x - 2, polynomial(minima_x) - 50),
             arrowprops=dict(facecolor='red', shrink=0.05))

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of 5th Degree Polynomial with Maxima and Minima')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
